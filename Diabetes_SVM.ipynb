```python
#importing all required libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset
df = pd.read_csv('/content/diabetes.csv')
```

```python
#printing first 5 rows of our data
df.head()
```

```python
#printing last 5 rows of our data
df.tail()
```

```python
#checking number of rows and columns in our dataset
df.shape
```

```python
#checking number of null/empty values
df.isnull().sum()
```

```python
#getting the statistical measure of our dataset
df.describe()
```

```python
#checking the total no of each type of outcomes, 1 for diabetic and 0 for non-diabetic
df['Outcome'].value_counts()
```

```python
#checking mean values of all the variables by grouping them for two outcomes
df.groupby('Outcome').mean()
```

```python
#separating the data and outcome
x = df.drop(columns = 'Outcome',axis=1)
y = df['Outcome']
```

```python
print(x)
```

```python
print(y)
```

```python
#Data Standardization
scaler = StandardScaler()
scaler.fit(x)
standardized_data = scaler.transform(x) #scaler.fit_transform(x)
print(standardized_data)
```

```python
x = standardized_data
y = df['Outcome']
```

```python
#Train-Test splitting of data
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, stratify=y, random_state=33)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#selection of model
model = svm.SVC(kernel='linear')
```

```python
#Training our SVM
model.fit(x_train,y_train)
```

```python
#Model Evaluation -> Finding Accuracy Score
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
print("Training data accuracy :",training_data_accuracy)
x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction,y_test)
print("Testing data accuracy :",testing_data_accuracy)
```

MAKING A DIABETES PREDICTIVE SYSTEM

```python
input_data = (1,85,66,29,0,26.6,0.351,31)

#Converting the input data array into a numpy array
np_data = np.asarray(input_data)

#reshaping the data
np_reshape = np_data.reshape(1,-1)

#standardizing the data
np_std = scaler.transform(np_reshape)

prediction = model.predict(np_std)
if(prediction[0]==1):
  print("The person is diabetic")
else:
  print("The person is non-diabetic")
```
